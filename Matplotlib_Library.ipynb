# Basic graph in Matplotlib
Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Month data (X-axis)
months = np.array([1, 2, 3, 4, 5])
# Monthly Sale (Y-axis)
sales = np.array([100, 150, 120, 200, 180])

# Create line plot
plt.plot(months, sales)
plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sale")
plt.show()
```

Bar Charts

```python
products = ['Rice', 'Daal', 'Oil']
stock = [50, 30, 15]

plt.bar(products, stock, color='red')
plt.title("Product stock")
plt.xlabel("products")
plt.ylabel("stock")
plt.show()
```

Scatter Plots

```python
hours_studied = [2, 4, 6, 8, 10]
exam_scores = [60, 75, 80, 95, 99]

plt.scatter(hours_studied, exam_scores, color='red')
plt.title("Hour studied vs Exam Scores")
plt.show()
```

Histograms

```python
customer_ages = [25, 30, 35, 35, 40, 40, 40, 50, 55, 60]

plt.hist(customer_ages, bins=5, color='orange')
plt.title("Age range")
plt.show()
```

Pie Charts

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
plt.pie(data, labels=cars)
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y

ax.scatter(x, y, z, c=c, cmap='magma')
ax.set_title('3D Line Plot')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y  # Color array based on x and y

ax.scatter(x, y, z, c=c)
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x**2 + y**3)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='magma', edgecolor='red')
ax.set_title('Surface Plot')
plt.show()
```

```python
def fun(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
X, Y = np.meshgrid(x, y)
Z = fun(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='magma', alpha=0.8)

ax.set_title('3D Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

# Customization in Matplotlib
Adding markers in plot

```python
import matplotlib.pyplot as plt

# Month data (X-axis)
x = np.array([1, 2, 3, 4, 5])
# Monthly Sale (Y-axis)
y = np.array([100, 150, 120, 200, 180])

plt.plot(x, y, 'red', marker='v', linestyle=':', label='Data Points')
plt.grid(axis='both', linestyle='-', alpha=0.6)
plt.figure(figsize=(10, 5))
plt.title('Simple Plot with Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(1, 11)
markers = ['o', 's', '^', 'v', 'D', '*', '+', 'x']

for i, marker in enumerate(markers):
    plt.plot(x, [i*2]*10, marker=marker, linestyle='')

plt.title('Different Matplotlib Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Visualize line style

```python
import matplotlib.pyplot as plt

x = [1, 3, 2, 9, 8]
plt.plot(x,'--')

plt.title('Line Style', fontsize=100, color='b')
plt.xlabel('X-axis', fontsize=100, color='r')
plt.ylabel('X-axis', fontsize=100, color='g')
plt.show()
```
